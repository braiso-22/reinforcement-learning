```python
import gymnasium as gym
print(gym.__version__)
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create the environment
env = gym.make('MountainCar-v0')
```

```python
print(env.reset())
print(env.observation_space)
```

```python
def discretizar(valor):
    retorno = ((valor - env.observation_space.low) /\
                (env.observation_space.high - env.observation_space.low )) * 20
    return tuple(retorno.astype(int))

q_table = np.zeros((20, 20, 3), dtype=np.float16)
```

```python
def bellman_equation(q_sa, recompensa, nuevo_estado, learning_rate, gamma):
    return q_sa + learning_rate * (recompensa + gamma * np.max(q_table[nuevo_estado]) - q_sa)
```

```python
episodios = range(0, 10000)

ratio_exploracion = 0.3
for episodio in episodios:
    recompensa_total = 0
    estado, info = env.reset()
    estado = discretizar(estado)
    done = False
    while not done:
        if random.uniform(0, 1) < ratio_exploracion:
            accion = env.action_space.sample()
        else:
            accion = np.argmax(q_table[estado])

        proximo_estado, recompensa, done, truncated, info = env.step(accion)
        if truncated:
            break
        recompensa_total += recompensa
        proximo_estado = discretizar(proximo_estado)
        q_table[estado][accion] = bellman_equation(
            q_sa=q_table[estado][accion],
            recompensa=recompensa,
            nuevo_estado=proximo_estado,
            learning_rate=0.1,
            gamma=0.9
        )
        estado = proximo_estado
    print("Episodio: {}, Recompensa: {}".format(episodio, recompensa_total))


env.close()
```
